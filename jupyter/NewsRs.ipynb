```python
import json
```

```python
page_num = 1
l=[]
file_path = 'p{}.json'.format(page_num)
for i in range(1,7):    
    with open(file_path, 'r') as f:
        d = json.load(f)
        l += d['data']['objects']
        
print(len(l))
```

```python
l
```

```python
mydata = []
for item_dict in l:
    key = item_dict['title_en']
    value = item_dict['summary']
    mydata.append((key,value))
len(mydata)
with open('mydata.txt','w') as out:
    for id,item in enumerate(mydata):
        out.write(str(id)+'\t'+item[0]+"\t"+item[1]+'\n')
```
