```python
train_data.plot(kind="scatter", x="item_price_level", y="item_sales_level")
```

```python
sns.jointplot(x="item_price_level", y="item_sales_level", data=train_data, size=5)
```

```python
# 将“是否完成交易”这一信息加入上面的图中
sns.FacetGrid(train_data, hue="is_trade", size=5) \
   .map(plt.scatter, "item_price_level", "item_sales_level") \
   .add_legend()
```

```python
# 抽样数据做可视化
sample_data = train_data.sample(n=20000) 
sns.violinplot(x=sample_data["item_collected_level"])
```

```python
sns.violinplot(x=sample_data["item_price_level"])
```

```python
sns.boxplot(x="is_trade", y="item_collected_level", data=sample_data)
```

```python
sample_data_1000 = train_data.sample(1000)
ax = sns.violinplot(x="item_price_level", y="item_pv_level", hue="is_trade",data=sample_data, palette="muted", split=True)
```

```python
sns.violinplot(y="item_price_level", x="is_trade", data=sample_data_1000, size=6)
```

```python
#价格分布　成交、未成交差不多 --抽样数据
sns.FacetGrid(sample_data, hue="is_trade", size=6) \
   .map(sns.kdeplot, "item_price_level") \
   .add_legend()
```

```python
＃　train_data　　item_pv_level分布
sns.FacetGrid(train_data, hue="is_trade", size=6) \
       .map(sns.kdeplot, "item_pv_level") \
   .add_legend()
```
